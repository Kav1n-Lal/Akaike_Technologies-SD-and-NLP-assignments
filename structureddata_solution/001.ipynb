```python
X = c.iloc[:, :-1].values
y = c.iloc[:, -1].values
X
```

## Splitting the dataset into train and test sets

```python
# Separating into training and testing set
X_train, X_test, y_train, y_test = train_test_split(
    X,  # predictors
    y,  # target
    test_size=0.3,  # percentage of obs in test set
    random_state=102023,
)  # seed to ensure reproducibility

X_train.shape, X_test.shape
```

## Checking whether target variable is balanced or not

```python
pd.Series(y_train).value_counts().plot.bar(figsize=(5,5),color='purple')
plt.show()
```

**The 'result' column which depicts whether the patient has taken the TARGET DRUG or not is our dependent variable and it is imbalanced**

## Applying SMOTE to oversample minority class

```python
# Balancing the target-only in the training set

sm = SMOTE(
    sampling_strategy='auto', # samples only the minority class
    random_state=102023,  # for reproducibility
    k_neighbors=10
    )
    

X_train_sm, y_train_sm = sm.fit_resample(X_train, y_train)
```

```python
# set up the scaler
scaler_sm = StandardScaler()

X_train_sm=scaler_sm.fit_transform(X_train_sm) # fitting the scaler and transforming the train set, 
                                               # to learn the parameters

X_test_sm= scaler_sm.transform(X_test)         # transforming the test set using the
                                               # learned parameters from the train set
```

## Checking for imbalance again

```python
pd.Series(y_train_sm).value_counts().plot.bar(figsize=(5,5),color='purple')
plt.show()
```

**Now the 'result' column-our target variable has been balanced**

## Defining classifier function

```python
def classifier(model_name,ml_model,X_train_set,y_train_set,X_test_scaled):
    model = ml_model
    model.fit(X_train_set,y_train_set)
    y_pred =model.predict(X_test_scaled)
    

    print('Train set')
    pred = model.predict_proba(X_train_set)
    #print(pred[:5])
    print(
        f'{model_name}',"roc-auc: {}".format(roc_auc_score(y_train_set, pred[:, 1])))
    #plot_confusion_matrix(model,X_train_set,y_train_set)
    print('Test set')
    pred = model.predict_proba(X_test_scaled)
    print(
        f'{model_name}',"roc-auc: {}".format(roc_auc_score(y_test, pred[:, 1])),'\n')
    print('accuracy_score:',accuracy_score(y_test,y_pred))
    print('\n')
    print(classification_report(y_test,y_pred))
    plot_confusion_matrix(model,X_test_scaled,y_test)
   
    
```

```python
classifier('logistic_regression_smote',LogisticRegression(random_state=102023),X_train_sm,y_train_sm,X_test_sm)
```

```python
rf_sm=RandomForestClassifier(random_state=23)
n_estimators=[34,35]
max_features= [4,5]
max_depth= [3,4,5]
bootstrap = [False]
oob_score = [False]
param_grid_rf = {'n_estimators':n_estimators,
                 'max_depth':max_depth,
             'max_features':max_features,
             'bootstrap':bootstrap,
             'oob_score':oob_score} 

grid_rf_sm = GridSearchCV(estimator = rf_sm,
                            param_grid = param_grid_rf,
                            scoring = 'roc_auc',
                            cv = 5,
                            )
```

```python
grid_rf_sm.fit(X_train_sm, y_train_sm)
print(grid_rf_sm.best_score_)
print(grid_rf_sm.best_params_)
```

```python
y_pred= grid_rf_sm.predict(X_test_sm)
print('accuracy_score: ',accuracy_score(y_test,y_pred))
pred = grid_rf_sm.predict_proba(X_train_sm)
print('Train set','\n',
    'Random_forest_classifier_smote roc-auc: {}'.format(roc_auc_score(y_train_sm, pred[:, 1])))


pred = grid_rf_sm.predict_proba(X_test_sm)
print('Test set','\n',
    'Random_forest_classifier_smote roc-auc: {}'.format(roc_auc_score(y_test, pred[:, 1])),'\n')
print(classification_report(y_test,y_pred))
plot_confusion_matrix(grid_rf_sm,X_test_sm,y_test)
```
