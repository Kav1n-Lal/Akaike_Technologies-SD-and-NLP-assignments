### I used the code below which:
<font color='green'> Groups the dataset based on 'Patient-Uid', takes the list of 'Incident' and 'Dates' of the same 'Patient-Uid' and records the number of times the patient has taken each drug along with the TARGET DRUG. </font>

```python
data=[]
DATE_DIFF1=[]
DATE_DIFF2=[]
k=0
u=list(df['Patient-Uid'])
inci=list(df['Incident'])
#df['diff_days'] = (df['end_date'] - df['start_date']) / np.timedelta64(1, 'D')
for id in uid_unique:
    r=[]
    a=df[df['Patient-Uid']==id]
    a1=list(a['Incident'])
    a2=list(a['Date'])
    r.append(id)

    for t in incident_unique:
        if t not in a1:
            r.append(0)
        elif (t in a1) and (t!='TARGET DRUG'):
            r.append(a1.count(t))
        elif (t in a1) and (t=='TARGET DRUG'):
            r.append(a1.count(t))
            e=a1.index('TARGET DRUG')
            d1=a2[e-1]
            d2=a2[e]
            DATE_DIFF1.append(d1)                
            DATE_DIFF2.append(d2)
        elif 'TARGET DRUG' not in a1:
            DATE_DIFF1.append(a2[-1])                
            DATE_DIFF2.append(a2[-1])
    data.append(r)
    k+=1
    print(f'{k} records extracted')
          
                
                
```
